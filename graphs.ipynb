# Evaluating the fitness of the robots

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('fitness_history.csv')

plt.figure(figsize=(10, 6))
plt.plot(df['Generation'], df['Avg_Internal_Score_Per_Match'], label='Internal Fitness')
plt.plot(df['Generation'], df['Avg_External_Score_Per_Match'], label='External Fitness')

plt.xlabel('Generation')
plt.ylabel('Score')
plt.title('Internal vs External Fitness per Generation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('fitness_history.csv')

WINDOW_SIZE = 5  # Higher values make the line smoother (e.g., 5, 10, 20)

df['Internal_Smooth'] = df['Avg_Internal_Score_Per_Match'].rolling(window=WINDOW_SIZE).mean()
df['External_Smooth'] = df['Avg_External_Score_Per_Match'].rolling(window=WINDOW_SIZE).mean()

plt.figure(figsize=(12, 7))

plt.plot(df['Generation'], df['Avg_Internal_Score_Per_Match'], color='blue', alpha=0.2, linewidth=1)
plt.plot(df['Generation'], df['Avg_External_Score_Per_Match'], color='orange', alpha=0.2, linewidth=1)

plt.plot(df['Generation'], df['Internal_Smooth'], color='blue', label=f'Internal Fitness', linewidth=2)
plt.plot(df['Generation'], df['External_Smooth'], color='orange', label=f'External Fitness', linewidth=2)

plt.xlabel('Generation')
plt.ylabel('Score')
plt.title('Internal vs External Fitness (Smoothed)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
